```python
from langchain_openai import OpenAIEmbeddings
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
endpoint = "https://models.github.ai/inference"
model_name = "text-embedding-3-large"
```

```python
from google.colab import userdata
import os
github_token = userdata.get('GITHUB_TOKEN')
os.environ['GITHUB_TOKEN'] = github_token
```

```python
embeddings = OpenAIEmbeddings(
    base_url = endpoint,
    api_key=github_token,
    model=model_name,
    dimensions=300
)
```

```python
documents = [
    'Virat Kohli is one of cricket‚Äôs greatest modern-day batters, known for his fierce intensity and unmatched consistency across all formats. He led India to multiple ICC titles and holds the record for most centuries in ODI history, surpassing even Sachin Tendulkar',
    'MS Dhoni, India‚Äôs ‚ÄúCaptain Cool,‚Äù revolutionized leadership with his calm demeanor and tactical brilliance, guiding India to three ICC trophies. Beyond cricket, he built a ‚Çπ800-crore business empire and remains a beloved icon through his legacy with Chennai Super Kings and Dhoni Entertainment',
    'Shubman Gill, India‚Äôs rising star and current Test captain, is known for his elegant strokeplay and record-breaking double centuries across formats. From U19 World Cup hero to Gujarat Titans skipper, he‚Äôs become the face of India‚Äôs next-gen cricket revolution',
    'Rohit Sharma, India‚Äôs ‚ÄúHitman,‚Äù is the only player with three double centuries in ODIs and holds the record for the highest individual ODI score‚Äîa jaw-dropping 264. üáÆüá≥ As captain, he led India to two consecutive ICC titles (T20 World Cup 2024 & Champions Trophy 2025), cementing his legacy as one of cricket‚Äôs most successful leaders'
]
```

```python
doc_embeddings = embeddings.embed_documents(documents)
```

```python
query = "Tell me about Rohit Sharma"
```

```python
query_embedding = embeddings.embed_query(query)
```

```python
print(doc_embeddings)
```

```python
print(len(doc_embeddings))
```

```python
print(len(doc_embeddings[0]))
```

```python
result = cosine_similarity([query_embedding], doc_embeddings)
```

```python
result[0]
```

## Query was about Rohit Sharma and hence last document shows 0.65 - more similar

```python
print("find the relevant document")
updated_result = list(enumerate(result[0]))
updated_result
```

```python
sorted_result = sorted(updated_result, key=lambda x: x[1])
sorted_result
```

```python
print("Most similar is the last one")
sorted_result[-1][0]
```
